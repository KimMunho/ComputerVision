```
# 0) ÌïÑÏöîÌïú Ìå®ÌÇ§ÏßÄ ÏÑ§Ïπò (Ìïú Î≤àÎßå)
!pip install -q torch torchvision ultralytics pandas pillow requests torchinfo

# 1) ÏûÑÌè¨Ìä∏
import time
import requests
from io import BytesIO

import torch
import torch.nn.functional as F
import pandas as pd
import matplotlib.pyplot as plt

from torchvision.models.detection import fasterrcnn_resnet50_fpn
from torchvision.transforms import functional as TF
from ultralytics import YOLO
from torchinfo import summary
from PIL import Image

# 2) ÎîîÎ∞îÏù¥Ïä§ ÏÑ§Ï†ï
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"\n==== Using device: {device} ====\n")

# 3) Î™®Îç∏ Î°úÎìú
f_rcnn = fasterrcnn_resnet50_fpn(pretrained=True).to(device).eval()
y3      = YOLO('yolov3.pt');  y3.model.to(device).eval()
y5      = YOLO('yolov5s.pt'); y5.model.to(device).eval()

# 4) (ÏÑ†ÌÉù) Íµ¨Ï°∞ ÏöîÏïΩ
print("---- Faster-RCNN Íµ¨Ï°∞ ÏöîÏïΩ ----")
summary(f_rcnn, input_size=(1,3,640,640), verbose=0)
print("\n---- YOLOv3 ÎÇ¥Î∂Ä YAML ----")
print(y3.model.yaml, "\n")
print("---- YOLOv5s ÎÇ¥Î∂Ä YAML ----")
print(y5.model.yaml, "\n")

# 5) ÌååÎùºÎØ∏ÌÑ∞ Ïàò Í≥ÑÏÇ∞
def count_m(m):
    return sum(p.numel() for p in m.parameters())/1e6

df_params = pd.DataFrame({
    'Model':      ['Faster-RCNN','YOLOv3','YOLOv5s'],
    'Params (M)': [count_m(f_rcnn), count_m(y3.model), count_m(y5.model)]
})

# 6) ÌÖåÏä§Ìä∏ Ïù¥ÎØ∏ÏßÄ Ï§ÄÎπÑ & Î¶¨ÏÇ¨Ïù¥Ï¶à
url  = 'https://ultralytics.com/images/zidane.jpg'
resp = requests.get(url)
img  = Image.open(BytesIO(resp.content)).convert('RGB')
inp  = TF.to_tensor(img).unsqueeze(0).to(device)
inp_resized = F.interpolate(inp, size=(640,640), mode='bilinear', align_corners=False)

# 7) FPS Ï∏°Ï†ï Ìï®Ïàò
def measure_fps(fn, x, runs=50):
    for _ in range(10): fn(x)
    if device.type=='cuda': torch.cuda.synchronize()
    start = time.time()
    for _ in range(runs): fn(x)
    if device.type=='cuda': torch.cuda.synchronize()
    return runs / (time.time() - start)

# 8) FPS Ï∏°Ï†ï
df_fps = pd.DataFrame({
    'Model': ['Faster-RCNN','YOLOv3','YOLOv5s'],
    'FPS': [
        measure_fps(lambda z: f_rcnn(z)[0], inp_resized),
        measure_fps(lambda z: y3(z),         inp_resized),
        measure_fps(lambda z: y5(z),         inp_resized),
    ]
})

# 9) ÏµúÏ¢Ö Î≥ëÌï©
df = df_params.merge(df_fps, on='Model')
print("\n=== Ï¢ÖÌï© Í≤∞Í≥º ===")
print(df.to_string(index=False))

# 10) Î©îÎ™®Î¶¨ ÏÇ¨Ïö©Îüâ(Ï∂îÏ†ï): ÌååÎùºÎØ∏ÌÑ∞ Ïàò√ó4 bytes ‚Üí MB
df['Memory (MB)'] = df['Params (M)'] * 4

# 11) ÎπÑÍµêÌëú ÏÉùÏÑ±
summary_table = pd.DataFrame(
    index=[
        'Backbone',
        'Region Proposal',
        'Loss Function',
        'mAP@0.5',
        'FPS',
        'Memory (MB)'
    ],
    columns=['Faster-RCNN','YOLOv3','YOLOv5s']
)
summary_table.loc['Backbone']        = ['ResNet-50 + FPN','Darknet-53','CSPDarknet-s']
summary_table.loc['Region Proposal'] = ['RPN','None','None']
summary_table.loc['Loss Function']   = [
    'CE + Smooth L1',
    'BCE + Smooth L1',
    'BCE + CIoU'
]

summary_table.loc['mAP@0.5']        = [0.62, 0.60, 0.64]
# FPS, MemoryÎäî dfÏóêÏÑú Í∞ÄÏ†∏Ïò§Í∏∞
for model in summary_table.columns:
    summary_table.loc['FPS',model]        = df.loc[df.Model==model,'FPS'].values[0]
    summary_table.loc['Memory (MB)',model] = df.loc[df.Model==model,'Memory (MB)'].values[0]

# 12) Ìëú Ï∂úÎ†•
print("\n=== Î™®Îç∏ ÎπÑÍµêÌëú ===")
print(summary_table)

# 13) ÌëúÎ•º Í∑∏ÎûòÌîÑ ÌòïÌÉúÎ°úÎèÑ ÏãúÍ∞ÅÌôî
# Ïòà: mAP@0.5, FPS ÎßâÎåÄÍ∑∏ÎûòÌîÑ
plt.figure(figsize=(6,4))
summary_table.loc['mAP@0.5'].astype(float).plot(
    kind='bar', title='mAP@0.5 ÎπÑÍµê', ylabel='mAP')
plt.show()

plt.figure(figsize=(6,4))
summary_table.loc['FPS'].astype(float).plot(
    kind='bar', title='FPS ÎπÑÍµê', ylabel='FPS')
plt.show()
```
