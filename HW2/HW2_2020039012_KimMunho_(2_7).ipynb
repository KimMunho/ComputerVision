```
# OpenCV를 사용하기 위한 초기 설정

import cv2
import numpy as np   # 넘파이 사용
import matplotlib.pyplot as plt   # matplotlib : 시각화라이브러리
from google.colab.patches import cv2_imshow    # colab에서 cv.imshow 사용

# 이미지 불러오기

from google.colab import drive
drive.mount('/content/drive')

img_path = '/content/drive/MyDrive/Colab Notebooks/ComputerVision/HW2/Fig0939(a).tif'
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

if img is None:
    print("Import Failed")

import matplotlib.pyplot as plt

# 원본 이미지
print("원본이미지\n")
cv2_imshow(img)

### 1. Sobel
# 뇌 구조 외곽 강조하며 공간이 비어있는지 아닌지 판단이 가능함

sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
sobel = cv2.magnitude(sobel_x, sobel_y)
sobel = cv2.convertScaleAbs(sobel)

print("sobel\n")
cv2_imshow(sobel)

### 2. Laplacian
# 뇌 구조의 외곽만 도출해냄

laplacian = cv2.Laplacian(img, cv2.CV_64F)
laplacian = cv2.convertScaleAbs(laplacian)

print("laplacian\n")
cv2_imshow(laplacian)

### 3. Canny (임계값 자동 조정도 가능)
# 뇌 구조의 전체적인 선을 동일한 두께감으로 표현

canny = cv2.Canny(img, threshold1=50, threshold2=150)

print("canny\n")
cv2_imshow(canny)

### 4. Morphological Gradient = Dilation - Erosion
# 뇌 구조의 전체적인 선을 다른 두께감으로 표현

kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
morph_gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)

print("morph_gradient\n")
cv2_imshow(morph_gradient)


```
