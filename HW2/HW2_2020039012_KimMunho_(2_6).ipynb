```
# OpenCV를 사용하기 위한 초기 설정

import cv2
import numpy as np   # 넘파이 사용
import matplotlib.pyplot as plt   # matplotlib : 시각화라이브러리
from google.colab.patches import cv2_imshow    # colab에서 cv.imshow 사용

# 이미지 불러오기

from google.colab import drive
drive.mount('/content/drive')

img_path = '/content/drive/MyDrive/Colab Notebooks/ComputerVision/HW2/Fig0905(a).tif'
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
_, bin_img = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

if img is None:
    print("Import Failed")

import matplotlib.pyplot as plt

# 원본 이미지
print("원본이미지\n")
cv2_imshow(bin_img)

# Kernel 정의
kernel1 = cv2.getStructuringElement(cv2.MORPH_RECT, (1, 3)) # 가장 얇은 두께
kernel2 = cv2.getStructuringElement(cv2.MORPH_RECT, (8, 3))  # 중간 두께
kernel3 = cv2.getStructuringElement(cv2.MORPH_RECT, (15, 3))  # 가장 넓은 두께

# kernel1 크기의 선분 제거
eroded1= cv2.erode(bin_img, kernel1, iterations=1)
print("kernel1 크기의 선분 제거\n")
cv2_imshow(eroded1)

# kernel2 크기의 선분 제거
eroded2 = cv2.erode(bin_img, kernel2, iterations=1)
print("kernel2 크기의 선분 제거\n")
cv2_imshow(eroded2)

# kernel3 크기의 선분 제거
eroded3 = cv2.erode(bin_img, kernel3, iterations=1)
print("kernel3 크기의 선분 제거\n")
cv2_imshow(eroded3)


```
