```
# OpenCV를 사용하기 위한 초기 설정

import cv2
import numpy as np   # 넘파이 사용
import matplotlib.pyplot as plt   # matplotlib : 시각화라이브러리
from google.colab.patches import cv2_imshow    # colab에서 cv.imshow 사용

# 이미지 불러오기

from google.colab import drive
drive.mount('/content/drive')

img_path = '/content/drive/MyDrive/Colab Notebooks/ComputerVision/HW2/Fig0938(a).tif'
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

if img is None:
    print("Import Failed")

import matplotlib.pyplot as plt

# 원본 이미지
print("원본이미지\n")
cv2_imshow(img)

# 가운데 흰색 연기 외의 부분을 노이즈로 판단

### 1. Morphological Smoothing (열기 후 닫기)
# 기준을 가지고 노이즈를 제거할 수 있으며 다소 딱딱한 이미지가 생성된다 (다소 왜곡된 형태)

kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 15))
morph_smoothed = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
morph_smoothed = cv2.morphologyEx(morph_smoothed, cv2.MORPH_CLOSE, kernel)


print("morph_smoothed\n")
cv2_imshow(morph_smoothed)

### 2. Smoothing Filter (가우시안 블러)
# 전체적으로 흐려지는 형태로 필터링 가능하며 부드러운 이미지 생성 가능

gaussian_blur = cv2.GaussianBlur(img, (11, 11), sigmaX=10)

print("gaussian_blur\n")
cv2_imshow(gaussian_blur)
```
