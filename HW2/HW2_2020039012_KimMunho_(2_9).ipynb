```
# OpenCV를 사용하기 위한 초기 설정

import cv2
import numpy as np   # 넘파이 사용
import matplotlib.pyplot as plt   # matplotlib : 시각화라이브러리
from google.colab.patches import cv2_imshow    # colab에서 cv.imshow 사용

# 이미지 불러오기

from google.colab import drive
drive.mount('/content/drive')

img_path = '/content/drive/MyDrive/Colab Notebooks/ComputerVision/HW2/Fig0940(a).tif'
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)


if img is None:
    print("Import Failed")

import matplotlib.pyplot as plt

# 원본 이미지

print("원본이미지\n")
cv2_imshow(img)

### Ostu 이진화
# 사라지는 쌀알이 생기며 노이즈가 다소 발생
_, bin_img = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

print("Ostu 이진화\n")
cv2_imshow(bin_img)


### Top-hat 변환 후 Otsu 이진화
# 사라지는 쌀알이 매우 적어졌으며 노이즈 또한 매우 작아짐

kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (50, 30))

# Top-hat 변환: 원본 - 열기(Opening)
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)

# Otsu 이진화 적용
_, otsu_tophat = cv2.threshold(tophat, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

print("Top-hat 변환 후 Otsu 이진화\n")
cv2_imshow(otsu_tophat)

```
