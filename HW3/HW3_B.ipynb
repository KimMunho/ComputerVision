```
!pip -q install torch torchvision scikit-learn matplotlib seaborn tqdm tabulate

import numpy as np, torch, torch.nn as nn
import matplotlib.pyplot as plt, seaborn as sns, random
import pandas as pd, tabulate, time, os
from torch.utils.data import DataLoader, TensorDataset
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from tqdm import tqdm

# ◇ 실험 파라미터 ────────────────────────────────
SEED          = 42
EPOCHS        = 300
BATCH_SIZE    = 64
LR            = 0.05
MOMENTUM      = 0.9
DEVICE        = "cuda" if torch.cuda.is_available() else "cpu"
HIDDEN_LAYERS = [512, 256, 128]

ACTS = {
    "ReLU"      : lambda: nn.ReLU(inplace=True),
    "LeakyReLU" : lambda: nn.LeakyReLU(0.1, inplace=True),
    "Sigmoid"   : lambda: nn.Sigmoid()
}

# 캡처할 에폭(0-based) 정의: 10, 150, 300
CAP_EPOCHS = [9, 149, 299]
# captures[act][epoch] = numpy array of Layer-1 activations
captures = {act: {ep: None for ep in CAP_EPOCHS} for act in ACTS}

torch.manual_seed(SEED); np.random.seed(SEED); random.seed(SEED)

# ◇ 데이터셋 ─────────────────────────────────────
X, y = make_moons(300, noise=0.25, random_state=SEED)
X = X.astype(np.float32);  y = y.astype(np.int64)
Xtr, Xte, ytr, yte = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=SEED)

tr_loader = DataLoader(
    TensorDataset(torch.from_numpy(Xtr), torch.from_numpy(ytr)),
    batch_size=BATCH_SIZE, shuffle=True)
te_loader = DataLoader(
    TensorDataset(torch.from_numpy(Xte), torch.from_numpy(yte)),
    batch_size=BATCH_SIZE, shuffle=False)

# ◇ 모델 정의 ───────────────────────────────────
class MLP(nn.Module):
    def __init__(self, d_in, hidden, d_out, act):
        super().__init__()
        layers, prev = [], d_in
        for h in hidden:
            layers.extend([nn.Linear(prev, h), act()])
            prev = h
        layers.append(nn.Linear(prev, d_out))
        self.net = nn.Sequential(*layers)
    def forward(self, x): return self.net(x)

def small_init(m):
    if isinstance(m, nn.Linear):
        nn.init.normal_(m.weight, 0.0, 0.01)
        nn.init.zeros_(m.bias)

# ◇ 학습·평가 루프 ───────────────────────────────
def run(act_name):
    model = MLP(2, HIDDEN_LAYERS, 2, ACTS[act_name]).to(DEVICE)
    model.apply(small_init)
    optim  = torch.optim.SGD(
        model.parameters(), lr=LR, momentum=MOMENTUM, nesterov=True)
    loss_f = nn.CrossEntropyLoss()
    hist, grads = {"loss":[], "acc":[]}, []

    for epoch in range(EPOCHS):
        # — 학습 단계
        model.train()
        for xb, yb in tr_loader:
            xb, yb = xb.to(DEVICE), yb.to(DEVICE)
            loss = loss_f(model(xb), yb)
            optim.zero_grad(); loss.backward(); optim.step()

        # — 평가 단계
        model.eval(); tot=cor=lsum=0
        with torch.no_grad():
            for xb, yb in te_loader:
                xb, yb = xb.to(DEVICE), yb.to(DEVICE)
                out = model(xb)
                lsum += loss_f(out, yb).item() * yb.size(0)
                cor  += (out.argmax(1)==yb).sum().item()
                tot  += yb.size(0)
        hist["loss"].append(lsum/tot)
        hist["acc"].append(cor/tot)

        # — gradient flow 기록
        grads.append([
            p.grad.detach().abs().mean().item()
            for p in model.parameters() if p.grad is not None
        ])

        # — 활성화 분포 캡처
        if epoch in CAP_EPOCHS:
            model.eval()
            with torch.no_grad():
                xb = torch.from_numpy(Xte).to(DEVICE)
                h0 = model.net[0](xb)    # 첫 Linear
                h1 = model.net[1](h0)    # Layer-1 activation
                captures[act_name][epoch] = h1.cpu().numpy().ravel()

    return model, hist, np.array(grads)

# ◇ Dead-ReLU 분석 ──────────────────────────────
def dead_stats(model):
    acts=[]
    def hook(_, __, out): acts.append(out.detach().cpu())
    hooks = [m.register_forward_hook(hook)
             for m in model.modules()
             if isinstance(m, (nn.ReLU, nn.LeakyReLU, nn.Sigmoid))]
    with torch.no_grad():
        for xb,_ in te_loader:
            model(xb.to(DEVICE)); break
    for h in hooks: h.remove()
    zeros = [((a==0).float().mean(0)).numpy() for a in acts]
    ratios = [100*(z==1).mean() for z in zeros]
    return zeros, ratios

# ◇ NaN 패딩 ─────────────────────────────────────
def pad_for_heatmap(arr_list, fill=np.nan):
    max_len = max(a.shape[0] for a in arr_list)
    mat = np.full((len(arr_list), max_len), fill, dtype=float)
    for i,a in enumerate(arr_list):
        mat[i, :a.shape[0]] = a
    return mat

# ◇ 메인 실행 ───────────────────────────────────
res, grads, models = {}, {}, {}
for act in ACTS:
    models[act], res[act], grads[act] = run(act)
    print(f"{act:9s} ▶ 최종 정확도 {res[act]['acc'][-1]*100:.2f}%")

# 1) Loss / Accuracy 그래프
plt.figure(figsize=(11,4))
for k,metric in enumerate(["loss","acc"]):
    plt.subplot(1,2,k+1)
    for act in ACTS:
        plt.plot(res[act][metric], label=act)
    plt.title(metric.upper()); plt.xlabel("Epoch"); plt.legend()
plt.tight_layout(); plt.show()


# 2) Dead-Neuron 히트맵
dead_mean = {}
for act in ACTS:
    z, ratio = dead_stats(models[act])
    dead_mean[act] = np.mean(ratio)
    print(f"{act} Dead-ReLU % per layer:", [f"{r:4.1f}%" for r in ratio])
    heat = pad_for_heatmap(z)
    sns.heatmap(heat, cmap="Reds", cbar=True, mask=np.isnan(heat),
                yticklabels=[f"L{i+1}" for i in range(len(z))])
    plt.title(f"Dead Map – {act}"); plt.xlabel("Neuron"); plt.show()

# 3) Gradient Flow 히트맵
for act in ACTS:
    g = grads[act]
    sns.heatmap(g.T, cmap="Blues",
                yticklabels=[f"L{i+1}" for i in range(g.shape[1])])
    plt.title(f"Gradient Flow – {act}")
    plt.ylabel("Layer"); plt.xlabel("Epoch")
    plt.show()

# 4) 초반·중반·후반 Layer-1 분포 비교
for act in ACTS:
    fig, axes = plt.subplots(1, len(CAP_EPOCHS), figsize=(15,4))
    for i, ep in enumerate(CAP_EPOCHS):
        data = captures[act][ep]
        sns.boxplot(x=data, ax=axes[i])
        axes[i].set_title(f"{act} @ Epoch {ep+1}")
    fig.suptitle(f"{act} Layer-1 분포 변화", fontsize=16)
    plt.tight_layout(rect=[0,0,1,0.94])
    plt.show()

# 5) 정량 비교표
df = pd.DataFrame({
    "Activation"      : list(ACTS.keys()),
    "Dead ReLU (%)"   : [dead_mean[a]           for a in ACTS],
    "Final Acc (%)"   : [res[a]["acc"][-1]*100  for a in ACTS],
    "Converged @Ep"   : [
        next((i for i,v in enumerate(res[a]["acc"]) if v>=0.95), "--")
        for a in ACTS]
})
print(tabulate.tabulate(df, headers="keys", showindex=False, floatfmt=".2f"))


```
